```python
#Need to setup a loop to iterate thru each team's data for each year
#first build a list of team's as found in pff's link
#should actually setup a dictionary instead to cover teams that changed names ex stl. rams
#nevermind, this isn't needed... they use updated team name even from before their switches
import time
teamList = ['arizona-cardinals','atlanta-falcons','baltimore-ravens','buffalo-bills','carolina-panthers','chicago-bears','cincinnati-bengals','cleveland-browns','dallas-cowboys','denver-broncos','detroit-lions','green-bay-packers','houston-texans','indianapolis-colts','jacksonville-jaguars','kansas-city-chiefs','los-angeles-chargers','los-angeles-rams','miami-dolphins','minnesota-vikings','new-england-patriots','new-orleans-saints','new-york-giants','new-york-jets','oakland-raiders','philadelphia-eagles','pittsburgh-steelers','san-francisco-49ers','seattle-seahawks','tampa-bay-buccaneers','tennessee-titans','washington-redskins']

#take me to page I want and find data needed
for teamName in teamList:
    for seasonYear in [x for x in range(2018,2019)]:
        driver.get("https://premium.profootballfocus.com/nfl/teams/" + teamName + "/summary?season=" + str(seasonYear) + "&weekGroup=REGPO#")
        time.sleep(5)
        #driver.implicitly_wait(15)
        weeks = driver.find_element_by_class_name("kyber-table-body__sticky-rows-container").find_elements_by_class_name('kyber-table-body__row')
        data = driver.find_element_by_class_name("kyber-table-body__scrolling-rows-container").find_elements_by_class_name('kyber-table-body__row')
        team = driver.find_element_by_class_name("kyber-filter-teams").find_element_by_class_name('kyber-filter-dropdown__toggle-text')
        
        #print(weeks)
        
        #data is setup in two sections within the table
        #NEED TO DEAL WITH EXTRA 'EMPTY' ROW THAT IS SPIT OUT in both lists...
        weekData = [week.text for week in weeks if week.text != '']
        gameData = [nugget.find_elements_by_class_name('kyber-table-body-cell') for nugget in data if nugget.text != '']
        team = team.text
        
        #print(weekData)
        
        weekList = []
        for week in weekData:
            newWeek = week.replace('\n@','')
            weekList.append(newWeek)
            
        gameList = []
        for game in gameData:
            for i in game:
                if i.text != '':
                    gameList.append(i.text)
                    
        #find what number we need for modular algebra
        numCols = len(gameList)/len(weekData)
        #print(numCols)
        
        weeksPlayed = len(weekList)
        #create lists for columns of the data frame
        allLists = [[] for i in range(0, int(weeksPlayed))]
        
        
        
        listNumber = 0
        #each list in allLists will be a row in our pandas df
        #iterate thru each list and push in the correct information to each from gameList

        for i in range(0,len(gameList)):
            #if the information from gameList is in the first row of data, 
            #add it to the first list in allLists, if not, move on to next list
            #when iterating thru, once we reach the 17th element in this case, we move on to filling the next list in allLists
            if i%numCols < numCols-1:
                allLists[listNumber].append(gameList[i])
            elif i%numCols == numCols-1:
                allLists[listNumber].append(gameList[i])
                listNumber += 1
        
        #create lists for a column of which team/season it is:
        teamList = [team] * len(weekList)
        seasonList = [seasonYear] * len(weekList)
        
        
        #need to combine each team list element to allLists
        for i in range(0,len(allLists)):
            allLists[i].insert(0,teamList[i])

        #need to combine each week list to allLists
        for i in range(0,len(allLists)):
            allLists[i].insert(0,weekList[i])
            allLists[i].insert(0,seasonList[i])

            
        seasonDf = pd.DataFrame(allLists, columns= tableVars)
        
        mainDf = mainDf.append(seasonDf, ignore_index=True)
```
