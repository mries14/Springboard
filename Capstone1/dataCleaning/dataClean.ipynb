```python
# Now clean up the game data from pro-football-reference
gameDf = pd.read_csv('C:\\Users\\Michael\\Desktop\\Springboard\\Capstone1\\dataScrapers\\csvFiles\\gameData2.csv')
```

```python
# Turn overUnder into floats
gameDf['overUnder'] = gameDf['overUnder'].str.slice(0,4)
gameDf['overUnder'] = gameDf['overUnder'].astype(float)
gameDf.head()
```

```python
# Turn date column into datetime object
# Also need to figure out what to do with vegasLine column:
# Either turn into two separate columns (favorite and spread), or turn into two columns that has homeSpread and awaySpread
```

```python
# Turn date column into datetime object
gameDf['date'] = pd.to_datetime(gameDf['date'])
gameDf.head()
```

```python
# Create a 'result' column that determines whether a game went over or under the total
gameDf.loc[gameDf.totalScore > gameDf.overUnder, 'result'] = 'over'
gameDf.loc[gameDf.totalScore == gameDf.overUnder, 'result'] = 'push'
gameDf.loc[gameDf.totalScore < gameDf.overUnder, 'result'] = 'under'
gameDf
```

```python
# Rearrange columns so result is right after overUnder
cols = list(gameDf.columns)
cols = ['gameId','date','awayTeam','awayScore','homeTeam','homeScore','totalScore','overUnder','result','stadium','roof','surface',
 'vegasLine','referee','umpire','headLinesman','lineJudge','backJudge','sideJudge','fieldJudge']
gameDf = gameDf[cols]
gameDf.head()
```

```python
# Need to replace all old team names with newest team names for data consistency among all tables
# St. Louis Rams to LA, SD Chargers to LAC
gameDf['awayTeam'] = gameDf['awayTeam'].replace('St. Louis Rams', 'Los Angeles Rams')
#gameDf[gameDf['awayTeam'] == 'Los Angeles Rams']

gameDf['homeTeam'] = gameDf['homeTeam'].replace('St. Louis Rams', 'Los Angeles Rams')
#gameDf[gameDf['homeTeam'] == 'Los Angeles Rams']

gameDf['awayTeam'] = gameDf['awayTeam'].replace('San Diego Chargers', 'Los Angeles Chargers')
#gameDf[gameDf['awayTeam'] == 'Los Angeles Chargers']

gameDf['homeTeam'] = gameDf['homeTeam'].replace('San Diego Chargers', 'Los Angeles Chargers')
#gameDf[gameDf['homeTeam'] == 'Los Angeles Chargers']
```

```python
# Make surface column grass if the value contains grass, and turf otherwise
gameDf.loc[gameDf['surface'].str.contains('grass'), 'surface'] = 'grass'
gameDf.loc[np.logical_not(gameDf['surface'].str.contains('grass')), 'surface'] = 'turf'
gameDf['surface'].astype('category')
gameDf['surface'].value_counts()
```

```python
# Make roof column outdoors or dome 
gameDf.loc[gameDf['roof'].str.contains('outdoors'), 'roof'] = 'outdoors'
gameDf.loc[np.logical_not(gameDf['roof'].str.contains('outdoors')), 'roof'] = 'dome'
gameDf['roof'].astype('category')
gameDf['roof'].value_counts()
```

```python
# Next we will clean the pff data
pffDf = pd.read_csv('C:\\Users\\Michael\\Desktop\\Springboard\\Capstone1\\dataScrapers\\csvFiles\\pffData.csv')
```

```python
# First we will turn date into a datetime object
pffDf['season'] = pffDf['season'].astype(str)
```

```python
# Need to filter out all games in Jan/Feb because same season, but new year
pffDf.loc[pffDf['date'].str.contains('01/|02/'), 'lateGame'] = True
pffDf.loc[pffDf['lateGame'] != True, 'lateGame'] = False
```

```python
# turn season into int
pffDf.season = pffDf.season.astype(int)
```

```python
# Now create new 'year' column that is the correct year given when the game is
pffDf['year'] = np.where(pffDf['lateGame'] == True, pffDf['season'] + 1, pffDf['season'])
# Cast year back into a string so we can turn the correct dates into datetime objects
pffDf['year'] = pffDf['year'].astype(str)
```

```python
pffDf['date'] = pffDf[['year','date']].apply(lambda x: ' '.join(x), axis=1)
pffDf['date'] = pd.to_datetime(pffDf['date'])
```

```python
pffDf[45:60]
```

```python
# First need to separate the LA teams and NY teams for the team column manually
# Need to deal with this issue in the scraper at some point
pffDf.loc[3465:3683,'team'] = 'Los Angeles Chargers'
pffDf.loc[3684:3895,'team'] = 'Los Angeles Rams'
pffDf.loc[4780:4998,'team'] = 'New York Giants'
pffDf.loc[4999:5213,'team'] = 'New York Jets'
```

```python
# Collect team names from team column and opponent column.
# Make team name consistent with opponent team names
teamSet = set(pffDf['team'])
oppSet = set(pffDf['opp'])
teamList = list(teamSet)
oppList = list(oppSet)
teamList.sort()
oppList.sort()
zipTeams = zip(teamList, oppList)
teamDict = dict(list(zipTeams))
```

```python
# Now turn 'team' column into the values in our dictionary using
# the current 'team' value as the key. Use .map(dictionary)

pffDf['team'] = pffDf['team'].map(teamDict)
pffDf.head()
```

```python
# Turn weeks into numbers for playoff weeks
weekSetList = list(set(pffDf['week']))
# make a new set of weeks that turns playoff weeks into numbers based on weekSetList
newWeeks = ['20','17','18','1','21','9','7','4','2','6','10','11','14','15','3','12','13','16','19','5','8']
# Now zip and map old df weeks column to new one
zipWeeksDict = dict(list(zip(weekSetList,newWeeks)))
pffDf['week'] = pffDf['week'].map(zipWeeksDict)
```
