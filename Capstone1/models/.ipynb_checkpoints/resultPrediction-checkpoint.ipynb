```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import random
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import datetime
from sklearn.model_selection import train_test_split, KFold, cross_val_score
```

```python
gameTable = pd.read_csv('C://Users//Michael//Desktop//Springboard//Capstone1//dataCleaning//cleanedData//cleanedGameData.csv')
dvoaTable = pd.read_csv('C://Users//Michael//Desktop//Springboard//Capstone1//dataCleaning//cleanedData//cleanedDvoaData.csv', encoding='windows-1252')
pffTable = pd.read_csv('C://Users//Michael//Desktop//Springboard//Capstone1//dataCleaning//cleanedData//cleanedPffData.csv')
```

```python
# Need to make new pff tables that have all dates covered... Dates in pff table are off by one or two days
# Strategy is to create new df's that have all dates, then join the normal pff table to each of them by date,
# then merge all the tables together. All dates will be in the final table
gameTable['date'] = pd.to_datetime(gameTable['date'])
pffTable['date'] = pd.to_datetime(pffTable['date'])
pffB1 = pffTable.copy()
pffB1['date'] = pffB1['date'] - datetime.timedelta(days=1)
pffB2 = pffTable.copy()
pffB2['date'] = pffB2['date'] - datetime.timedelta(days=2)
pffF1 = pffTable.copy()
pffF1['date'] = pffF1['date'] + datetime.timedelta(days=1)
pffF2 = pffTable.copy()
pffF2['date'] = pffF2['date'] + datetime.timedelta(days=2)
```

```python
pffTable.head()
```

```python
pffF1.head()
```

```python
# Now join pffTable to each new df
t0 = pd.merge(gameTable,pffTable,how='inner',left_on=['date','awayTeam'], right_on=['date','team'])
t1 = pd.merge(gameTable,pffB1,how='inner',left_on=['date','awayTeam'], right_on=['date','team'])
t2 = pd.merge(gameTable,pffB2,how='inner',left_on=['date','awayTeam'], right_on=['date','team'])
t3 = pd.merge(gameTable,pffF1,how='inner',left_on=['date','awayTeam'], right_on=['date','team'])
t4 = pd.merge(gameTable,pffF2,how='inner',left_on=['date','awayTeam'], right_on=['date','team'])
t5 = pd.merge(gameTable,pffTable,how='inner',left_on=['date','homeTeam'], right_on=['date','team'])
t6 = pd.merge(gameTable,pffB1,how='inner',left_on=['date','homeTeam'], right_on=['date','team'])
t7 = pd.merge(gameTable,pffB2,how='inner',left_on=['date','homeTeam'], right_on=['date','team'])
t8 = pd.merge(gameTable,pffF1,how='inner',left_on=['date','homeTeam'], right_on=['date','team'])
t9 = pd.merge(gameTable,pffF2,how='inner',left_on=['date','homeTeam'], right_on=['date','team'])
```

```python
awayTable = t0.append([t1,t2,t3,t4],ignore_index=True)
homeTable = t5.append([t6,t7,t8,t9],ignore_index=True)
```

```python
awayTable = awayTable[['gameId','date','season','week','awayTeam']]
homeTable = homeTable[['gameId','date','season','week','homeTeam']]
```

```python
# Traansform week from dvoa table into correct format for analysis (add one)
dvoaTable['week'] = dvoaTable['week'] + 1
```

```python
# Next, join these tables with DVOA table
awayTable = awayTable.merge(dvoaTable, left_on=['season','week','awayTeam'], right_on=['season','week','team'])
homeTable = homeTable.merge(dvoaTable, left_on=['season','week','homeTeam'], right_on=['season','week','team'])

# Need to get superbowl data still; left out because dvoa weeks only go until 20
```

```python
awayTable = awayTable.rename(index=str, columns={'totalDvoa': 'awayTotalDvoa','offDvoa':'awayOffDvoa','offRank':'awayOffRank','defDvoa':'awayDefDvoa','defRank':'awayDefRank','stDvoa':'awayStDvoa','stRank':'awayStRank'})
homeTable = homeTable.rename(index=str, columns={'totalDvoa': 'homeTotalDvoa','offDvoa':'homeOffDvoa','offRank':'homeOffRank','defDvoa':'homeDefDvoa','defRank':'homeDefRank','stDvoa':'homeStDvoa','stRank':'homeStRank'})
```

```python
awayTable = awayTable.drop('team',axis=1)
homeTable = homeTable.drop('team',axis=1)
```

```python
# Now join awayTable and homeTable to gameTable
mainDf = gameTable.merge(awayTable, on=['gameId','date','awayTeam']).merge(homeTable, on=['gameId','date','season','week','homeTeam'])
```

```python
mainDf
```

```python
# Take out cols we don't want
mainDf = mainDf.loc[:,['gameId', 'date','season', 'week', 'stadium', 'roof', 'surface',
       'awayTeam', 'awayTotalDvoa', 'awayOffDvoa', 'awayDefDvoa', 'awayStDvoa',
       'homeTeam', 'homeTotalDvoa', 'homeOffDvoa', 'homeDefDvoa', 'homeStDvoa',
       'overUnder', 'totalScore', 'result']]
```

```python
mainDf
```

```python
mainDf = mainDf[mainDf['result'] != 'push']
```

```python
mainDf = mainDf[mainDf['week'] >= 3]
```

```python
#mainDf = mainDf[mainDf['stadium'].str.contains('Heinz')]
```

```python
# Need split data into two random samples for training/testing
# First, import train_test_split

```

```python
X_train, X_test, y_train, y_test = train_test_split(mainDf.iloc[:,:-1],mainDf.iloc[:,-1:], test_size=0.2,random_state=0)
```

```python
ID_col = ['gameId']
target_col = ['result','totalScore']
cat_cols = ['roof','surface']
num_cols= ['awayTotalDvoa','awayOffDvoa','awayDefDvoa','awayStDvoa','homeTotalDvoa','homeOffDvoa','homeDefDvoa','homeStDvoa','overUnder']
other_col=['date','season','week','stadium','awayTeam','homeTeam','Type'] #Test and Train Data set identifier
```

```python
# Create label encoders for categorical variables and split the data set
# to train & test, further split the train data set to Train and Validate

#create label encoders for categorical features
for var in cat_cols:
 number = LabelEncoder()
 X_train[var] = number.fit_transform(X_train[var].astype('str'))

for var in cat_cols:
 number = LabelEncoder()
 X_test[var] = number.fit_transform(X_test[var].astype('str'))
```

```python
#Target variable is also a categorical so convert it
y_train['result'] = number.fit_transform(y_train['result'].astype('str'))
y_test['result'] = number.fit_transform(y_test['result'].astype('str'))
```

```python
# Pass the imputed and dummy (missing values flags) variables into the
# modelling process. Use random forest to predict the class
features=list(set(list(X_train.columns))-set(ID_col)-set(target_col)-set(other_col))
```

```python
X_train = X_train[list(features)]
X_test= X_test[list(features)]
```

```python
X = X_train.append(X_test)
y = y_train.append(y_test)
```

```python
rf = RandomForestClassifier(n_estimators=101, random_state=10)
k_fold = KFold(n_splits=10)
cross_val_score(rf, X, y, cv=k_fold)
```

```python
#random.seed(100)
rf = RandomForestClassifier(n_estimators=101, random_state=0)
rf.fit(X_train, y_train)
#x_test = x_test.reshape(-1,1)
y_pred = rf.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))  
print(classification_report(y_test,y_pred))  
print(accuracy_score(y_test, y_pred))
```
