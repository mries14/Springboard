```python
# Read in qb stats files
qbStatsFiles = [filename for filename in glob.glob("C:/Users/Michael/Desktop/Springboard/Capstone2/csvFiles/DFS_Data/QB/stats/*.csv")]
```

```python
qbStatsFiles
```

```python
# Extract season and fix playoff weeks to match other tables

dfs = []
dfsPlayoffs = []

for i in range(int(len(qbStatsFiles)/2)):
    df = pd.read_csv(qbStatsFiles[i])
    df['season'] = qbStatsFiles[i][-8:-4]
    dfs.append(df)

# second half of list has playoff games, need to fix their weeks, then combine with first table    
for i in range(int(len(qbStatsFiles)/2),int(len(qbStatsFiles)/2)*2):
    dfPlayoffs = pd.read_csv(qbStatsFiles[i])
    dfPlayoffs['season'] = qbStatsFiles[i][-8:-4]
    dfsPlayoffs.append(dfPlayoffs)

    
for i in range(len(dfsPlayoffs)):
    dfsPlayoffs[i].loc[dfsPlayoffs[i].Week == 1, 'Week'] = 18
    dfsPlayoffs[i].loc[dfsPlayoffs[i].Week == 2, 'Week'] = 19
    dfsPlayoffs[i].loc[dfsPlayoffs[i].Week == 3, 'Week'] = 20
    dfsPlayoffs[i].loc[dfsPlayoffs[i].Week == 4, 'Week'] = 21


dfsPlayoffs[-1].head()
```

```python
qb = pd.concat(dfs,axis=0)
qb = qb.append(dfsPlayoffs)
qb.head()
```

```python
# Now fetch the salary data
# put salaries into a table and join them

qbSalariesFiles = [filename for filename in glob.glob("C:/Users/Michael/Desktop/Springboard/Capstone2/csvFiles/DFS_Data/QB/salaries/*.csv")]

dfsSalaries = []
dfsPlayoffsSalaries = []

for i in range(int(len(qbSalariesFiles)/2),int(len(qbSalariesFiles)/2)*2):
    df = pd.read_csv(qbSalariesFiles[i])
    df['season'] = qbSalariesFiles[i][-8:-4]
    dfsSalaries.append(df)

# second half of list has playoff games, need to fix their weeks, then combine with first table    
for i in range(int(len(qbSalariesFiles)/2)):
    dfPlayoffs = pd.read_csv(qbSalariesFiles[i])
    dfPlayoffs['season'] = qbSalariesFiles[i][-8:-4]
    dfsPlayoffsSalaries.append(dfPlayoffs)

    
for i in range(len(dfsPlayoffsSalaries)):
    dfsPlayoffsSalaries[i].loc[dfsPlayoffsSalaries[i].Week == 1, 'Week'] = 18
    dfsPlayoffsSalaries[i].loc[dfsPlayoffsSalaries[i].Week == 2, 'Week'] = 19
    dfsPlayoffsSalaries[i].loc[dfsPlayoffsSalaries[i].Week == 3, 'Week'] = 20
    dfsPlayoffsSalaries[i].loc[dfsPlayoffsSalaries[i].Week == 4, 'Week'] = 21
    dfsPlayoffsSalaries[i] = dfsPlayoffsSalaries[i].rename(columns={'FantasyPointsFanDuel':'Projection'})
    dfsSalaries[i] = dfsSalaries[i].rename(columns={'FantasyPointsFanDuel':'Projection'})
    
dfsPlayoffsSalaries[-1].head()
```

```python
qbSalaries = pd.concat(dfsSalaries,axis=0)
qbSalaries = qbSalaries.append(dfsPlayoffsSalaries)
len(qbSalaries)
```

```python
class QB:
    
    def __init__(self,name,table):
        self.name = name
        self.table = table[table['Name'] == name]
        
    def get_table(self):
        return self.table
    
    def corrMatrix(self,features):
        correlations = self.table[features].corr()
        return correlations.style.background_gradient(cmap='coolwarm')
    
    def scatter(self,var1,var2):
        plt.scatter(self.table[var1],self.table[var2])
        plt.title(self.name)
        plt.xlabel(var1)
        plt.ylabel(var2)
        plt.show()
```

```python
ad = QB('Matthew Stafford', qbs)
ad.corrMatrix(qbFeatures)
```
