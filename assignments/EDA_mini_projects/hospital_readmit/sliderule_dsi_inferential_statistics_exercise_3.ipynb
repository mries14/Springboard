# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
# Proportion of hospitals/facilities with between 100-300 discharges that have excess readmissions ratio > 1
df1 = clean_hospital_read_df[(clean_hospital_read_df['Number of Discharges'] < 300) & (clean_hospital_read_df['Number of Discharges'] > 100)]['Excess Readmission Ratio'].dropna()
P1 = sum(df1 > 1) / len(df1)
P1
```

```python
# Proportion of hospitals/facilities with between 100-300 discharges that have excess readmissions ratio > 1
df2 = clean_hospital_read_df[(clean_hospital_read_df['Number of Discharges'] > 300)]['Excess Readmission Ratio'].dropna()
P2 = sum(df2 > 1) / len(df2)
P2
```

If we are going to be proposing regulations on smaller hospitals, we should narrow down our definition of "smaller". Instead of including hospitals with < 300 discharges, we should change our definition to < 100 discharges. Our claim is that hospitals with between 100 - 300 discharges aren't significantly different than hospitals with > 300 discharges when it comes to the proportion of them having excess readmissions rate > 1.

Let P1 = the proportion of facilities that have between 100 and 300 discharges that have an excess readmission rate > 1, and let P2 = the proportion of facilities that have greater than 300 discharges that have an excess readmission rate > 1.

H0: P1 - P2 = 0 (no difference in proportion of readmissions rate > 1 between these two types of hospitals)

HA: P1 - P2 != 0 (there is a difference)

Then P1 = 0.51525, P2 = 0.49226, and P1 - P2 = 0.02299

Now we want to find the likelihood that we observe a value this extreme, assuming our null hypothesis (P1 - P2 = 0) is true. We want to find a Z-score, but we first need to compute the standard deviation of P1 - P2. Since we are assuming H0 is true, we can assume P1 = P2 = P, and compute the proportion P by assuming there is no difference in the proportion of P1 and P2. Therefore we can combine the two datasets, and use the total ratio the two datasets as P. 

```python
print(sum(df1 > 1))
print(sum(df2 > 1))
print(len(df1))
print(len(df2))
```

```python
P =(sum(df1 > 1) + sum(df2 > 1)) / (len(df1) + len(df2))
P
```

```python
# Now that we have our proportion, we can compute our standard deviation for the difference between the two groups,
# then find our z-score
stdDev = np.sqrt(((P*(1-P))/4885) + ((P*(1-P))/5365))
z = (0.02299 - 0) / stdDev
z
```

Using a p-value calculator, we find that the p-value = 0.020072. This exceeds our alpha level of 0.01, thus we fail to reject the null hypothesis that there is no difference between the two proportions. There is no sufficient evidence that the proportion of hospitals/facilities that have 100 to 300 discharges is significantly different than that of hospitals/facilities that have greater than 300 discharges when it comes to excess readmission rate > 1. Therefore, if we are going to propose a policy that forces facilites to upgrade resource allocation for quality care to continue operation, it shouldn't include hospitals of > 100 discharges.

When discussing statistical significance vs. practical significance, we note that a difference so large that it can be deemed as "real-life meaningful" rather than just "statistically significant" would be the distinction between the two. In our case, we don't get a statistically significant difference between proportions, so it wouldn't change our recommendation to the client. We would still recommend not imposing regulations based on number of discharges, unless doing so on a facility with < 100 discharges.
