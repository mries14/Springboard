```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_profiling
import datetime
```

```python
# First, read in the users data
users = pd.read_csv('takehome_users.csv',encoding='latin-1')
users.head()
```

```python
# explore the data and clean if needed
users.profile_report()
```

```python
# Turn creation_time variable into datetime object
users['creation_time'] = pd.to_datetime(users['creation_time'])
```

```python
engagement = pd.read_csv('takehome_user_engagement.csv')
engagement['time_stamp'] = pd.to_datetime(engagement['time_stamp'])
```

```python
engagement[engagement['user_id'] == 2]
```

```python
engagement.dtypes
```

```python
engagement['time_stamp'] = engagement['time_stamp'].dt.floor('d').astype(np.int64)
#sorting and remove duplicated days per users 
engagement = engagement.sort_values(['user_id', 'time_stamp']).drop_duplicates()

counter = engagement.groupby('user_id')['time_stamp'].rolling(window=3)
timeFrame = pd.to_timedelta((counter.max()- counter.min())).dt.days
```

```python
timeFrame.head()
```

```python
counter7days = timeFrame[timeFrame < 8].index.get_level_values('user_id').tolist()
adopted_users = list(set(counter7days))
```

```python
adopted_users.sort()
adopted_users[:10]
```

```python
# If user is adopted, indicate in 'adopted' column
users['adopted'] = users.isin(adopted_users)['object_id']
users[:10]
```

```python
# create new variable that indicates whether a user was invited by another or not
users.loc[np.isnan(users['invited_by_user_id']) == True, 'invited'] = False
users[:10]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
feature_cols = ['creation_source','opted_in_to_mailing_list','enabled_for_marketing_drip','org_id','invited']
X = users[feature_cols] # Features
y = users.adopted # Target variable

# Split the data into a training and test set
Xrf, Xtestrf, yrf, ytestrf = train_test_split(X,y,test_size=0.25,random_state=51)

# Now we want to predict status
# Setup table in way to do so:
target_col = ['adopted']
cat_cols = ['creation_source','opted_in_to_mailing_list','enabled_for_marketing_drip','org_id','invited']

# Create label encoders for categorical variables and split the data set
# to train & test, further split the train data set to Train and Validate

#create label encoders for categorical features
for var in cat_cols:
    number = LabelEncoder()
    Xrf[var] = number.fit_transform(Xrf[var].astype('str'))

for var in cat_cols:
    number = LabelEncoder()
    Xtestrf[var] = number.fit_transform(Xtestrf[var].astype('str'))
```

```python
clf = RandomForestClassifier()
# Fit the model on the trainng data.
clf.fit(Xrf, yrf)
# Print the accuracy from the testing data.
print(accuracy_score(clf.predict(Xtestrf), ytestrf))
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
y_pred=clf.predict(Xtestrf)
cnf_matrix = metrics.confusion_matrix(ytestrf, y_pred)
cnf_matrix
```

```python
print(classification_report(ytestrf, y_pred))
```

```python
features = Xrf.columns
importances = clf.feature_importances_
indices = np.argsort(-importances)[::-1]
plt.title('Feature Importances - Random Forest Classifier')
plt.barh(range(len(indices[:10])), importances[indices[:10]], color='b', align='center')
plt.yticks(range(len(indices[:10])), [features[i] for i in indices[:10]])
plt.xlabel('Relative Importance')
plt.show()
```

From the classification report above, we can see that our model is solid for predicting when a user will not adopt, but lacks when attempting to predict when a user will be consider an 'adopter'. Instead of focusing on why our model fails to predict when a user will be an adopter, we can focus on what factors lead to a user NOT adopting, and consider taking action on how to improve in this area. Judging from the feature importances chart above, we can see that the organization that the user belongs to is a key factor in our predictive model, and should be our main area of focus. We could focus on incentivizing users to create or join organizations that are more likely to lead to them becoming adopters of the product. We also could incentivize users within these organizations to adopt. A thorough analysis of groups who are more likely to contain adopting users should be a priority.
